```python
#Label encoding
#if we want to use label encoding then we use label encoder,for that we have to use sklearn package because if we want to use label encoder then we have to get that from sklearn package
from sklearn.preprocessing import  LabelEncoder
```

```python
df['CAEC']=le.fit_transform(df['CAEC'])   #in label encoding automatically the values are replaced based on the alphabetical order
df.head(3)    #we are getting the first 3 rows values in the dataset ,here in the output we have seen CAEC has 2,2,2 values like that
              #ex:hari ,srilekha,sailatha  here in this three words which will come first based on alphabetical order
              #    0      2        1            so,in this way the labelencoder works
```

```python
x=df.drop({'NObeyesdad'},axis=1)   #here we are dropping or removing the 16th column and stored in the variable 'x'
y=df['NObeyesdad']                 #here we are storing the 16th column in 'y'
```

#splitting the training and testing datasets


```python
from sklearn.model_selection import train_test_split   #In model_selection we are importing train_test_split module
```

```python
#we are having four values that is xtrain,xtest,ytrain,ytest
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=12)  #here x,y are initialized and for test_size=0.3 means we can take any percent i.e 0.1,0.4 etc and random_state=12 defines the data is shuffled 12 times and after shuffling the 12th time shuffling data we are getting in that 70% is training dataset,30% is test dataset
```

```python
xtrain    #here the data is shuffled like 2009,929,861 etc  and we get 70% training dataset like 1477 rows
```

```python
xtest     #here we are getting 30% test dataset like 634 rows
```

#building  and training the model

```python
from sklearn.tree import DecisionTreeClassifier   #for using decision tree
from sklearn.ensemble import RandomForestClassifier
```

```python
#initialising the model
dt= DecisionTreeClassifier()   #we have to check the performance if performance is good or not
                               #we are checking the accuracy,not every model is giving 100%  accuracy ,every model gives more than 70% accuracy
```

```python
#training the model
dt.fit(xtrain,ytrain)    #we are training the algorithm
```

```python
ypred=dt.predict(xtest)    #we are predicting the model or algorithm and to check the performance
print(ypred)
```

```python
ytest
```

#to check the classification report

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(ytest,ypred))     #here ytest is the actual value and ypred is the predicted value
                                              #in the output we get accuracy as 0.94 means 94 percent accuracy
```

```python
confusion_matrix(ytest,ypred)    #here what ever we are getting in the diagonal values are the exact values we have to consider here as accuracy and remaining are negligible
                                  #if we are getting more than 70% then it is not good accuracy like 78,4 it is not the better approach i.e accuracy
```
