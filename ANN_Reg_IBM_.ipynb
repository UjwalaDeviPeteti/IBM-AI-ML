```python
#encoding
#LbelEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()  #initialising the lib
```

```python
xtrain['State']=le.fit_transform(xtrain['State'])
xtrain['State']     #with the help of le.fit_transform we are transforming the object datatype into the int or float
```

```python
#Build an ANN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense     #with the help of dense we can create the 3 layers input,hidden,and output layers
```

```python
#initialising the seq model
reg_model=Sequential()
#adding the input layer to the model
reg_model.add(Dense(4,activation='relu'))      #we are taking 4 neurons
#adding the 1st hidden layer to the model
reg_model.add(Dense(64,activation='relu'))     #64 neurons
#adding the 2nd hidden layer to the model
reg_model.add(Dense(32,activation='relu'))    #32 neurons
#adding output layer
reg_model.add(Dense(1,activation='linear'))    #we are expecting output as 1 neuron  #for regression we use activation as linear and for classification we use sigmoid
```

```python
#compiling th eskeleton with loss and gradient descent
reg_model.compile(optimizer='adam',loss='mse')    #in gradient descent the errors are plotted in a graph so,that can be provided by optimizer so,in optimizer we are using the adam
```

```python
#training the model
reg_model.fit(xtrain,ytrain,batch_size=4,epochs=10 )  #epochs means in single go we cant test the model so we are giving values as 10 so,10 times we are checking this value is not default and ,branch_size means we are diving the total data into 4 parts )
```

```python
#testing the model
ypred=reg_model.predict(xtrain)    #predicting the values
```

```python
#comparing predicting value  with the actual value
comp=pd.DataFrame(ytrain)       #creating the dataframe
comp.columns=['Actual Value']   #changing the column names
comp['Predicted Values']=ypred
comp
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/iris.csv')
df.head()
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['species'].unique()    #to get the unique categories of species
```

```python
df['species'].value_counts()  #to get the no.of values that are associated with that unique feature
```

```python
#splitting the data
x=df.iloc[:,0:4]
x
```

```python
y = df['species']
y
```

```python
#splitting the data into training and test datasets
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=21)
```

```python
xtrain.shape,ytrain.shape,xtest.shape,ytest.shape
```

```python
le = LabelEncoder()
ytrain = le.fit_transform(ytrain)
ytest = le.transform(ytest)
```

```python
ytrain
```

```python
ytest
```

```python
model = Sequential()
model.add(Dense(4,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(26,activation='relu'))
model.add(Dense(3,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=10,batch_size=5,validation_data=(xtest,ytest))
```

```python
ypred = model.predict([[2.5,4.34,5.3,1.3]])
print(ypred)

```

```python
ypred = np.argmax(ypred)
output = ['Iris-setosa','Iris-versicolor','Iris-virginica']
output[ypred]
```
